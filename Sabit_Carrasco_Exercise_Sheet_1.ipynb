# 1. Aufgabe


```python
def fizz_buzz(n):
    if n%3 == 0:
        if n%5 == 0:
            return "FizzBuzz"
        else:
            return "Fizz"
    elif n%5 == 0:
        return "Buzz"
    else:
        return n
```

```python
# test cases
assert(fizz_buzz(15) == "FizzBuzz")
assert(fizz_buzz(3) == "Fizz")
assert(fizz_buzz(5) == "Buzz")
assert(fizz_buzz(2) == 2)
```

# 2. Aufgabe


```python
def running_mean(l):
    mean_i = np.zeros(len(l))
    for i in range(len(l)):
        mean_i[i]= np.mean(l[:i+1])
    return mean_i
```

```python
# test
l=[6,-5,-1,3,11,1,-8,7,4,9]
running_mean(l)
```

```python
# WG-Daten lesen
wg_data = pd.read_csv('input_data.csv', sep=';', names=['a', 'b', 'c', 'd'])
wg_data.head()
```

```python
# replacement dictionaries
stadtteile = {'Leipzig Neustadt-Neuschnefeld':0, 
              'Leipzig Zentrum-Nord':1,
              'Leipzig Zentrum-Sdost':2,
              'Leipzig Lindenau':3,
              'Leipzig Gohlis-Mitte':4,
              'Leipzig Zentrum':5,
              'Leipzig Sttteritz':6,
              'Leipzig Volkmarsdorf':7,
              'Leipzig Schleuig':8 }
zimmer_anzahl = {'1-Zimmer-Wohnung':0,
                 '2er WG':1,
                 '3er WG':2,
                 '4er WG':3,
                 '5er WG':4 }
```

# 4. Aufgabe


```python
class Node:
    def __init__(self, id, value='', alpha='', beta='', visited=False, children=[]):
        self.id = id
        self.value = value
        self.alpha = alpha
        self.beta = beta
        self.visited = visited
        self.children = children
        
    def get_children(self):
        return self.children
    
    def __repr__(self):
        return f"Node {self.id}"
```

```python
bob = Node(id=7, value="-8", alpha="", beta="", visited=False, children=[])
bob.value.items
```

```python
# Graph erstellen
bob = Node(id=7, value="-8", alpha="", beta="", visited=False, children=[])
helene = Node(id=6, value="5", alpha="896", beta="", visited=False, children=[bob])
beate = Node(id=5, value="13", alpha="", beta="1278", visited=False, children=[helene])
donald = Node(id=4, value="-4108", alpha="1", beta="3209", visited=False, children=[])
tim = Node(id=3, value="42", alpha="-inf", beta="inf", visited=False, children=[beate, donald])
gerald = Node(id=2, value="37", alpha="", beta="inf", visited=False, children=[])
prapra = Node(id=1, value="1258", alpha="4369", beta="40", visited=False, children=[gerald, tim])
```
